# 7) Write a function that takes a username and returns the full dictionary for that username

The function should take one argument and return the dictionary from `reviewers_dicts` that contains the given username in its `Username` field. If the username is not found, it should return `None`.

Name the function `find_user`

Test it by calling it with `krpz1113` and `rtm3422`

```python
def find_user(val):
    """Returns all the values in a users dictionary based on their username"""
    for x in reviewers_dicts:
        if val == x['Username']:  #If the input matches a known Username the dictionary will be returned
            return x
    return "None"  #If the input doesnt match a known Username None will be returned

print(find_user('krpz1113'))  #Passing a Username into the function 
print(find_user('rtm3422'))  #Passing a Username into the function
```

# 8) Write another function that takes a product id and returns the full dictionary for the associated product

Name the function `find_product`

# 9) Now use the functions you have just written to list a user's reviews

Write a *generator* function that takes a username and yields all the reviews the person with that username has written, one at a time.

Call the function to get a generator object and then iterate over it to list the reviews written by username `krpz1113`.

```python
def user_reviews(username):
    """Returns all the reviews in a users dictionary based on their username"""
    for product_dict in products_dicts:  #Iterating over the list of dictionaries
        reviews = product_dict["Reviews"]  #Assigning the reviews to a variable 
        for user_review in reviews:  #Nested for loop to loop over the dictionary within the list
            user = user_review["User"]  #Creating a new variable to store each user in
            if user == username:  #Storing only the user we input into the generator
                yield user_review  #Yielding the result back outside the generator

for user_review in user_reviews("krpz1113"):  #Iterating over our generator with the required input
    print(user_review)
```

```python
merger = pd.merge(reviews_pd, users_pd, how='left', on='Username')[['State', 'Score']]  #Joining two tables to access the required data

grouped_merger = merger.groupby("State")  #Grouping the data based on the state
mean_merger = grouped_merger.mean()  #Calculating the mean of the scores
mean_merger = mean_merger.reset_index()  #Resetting the index 
               
display(mean_merger.head(5))  #Displaying the first 5 results
```

```python
import matplotlib.pyplot as plt  #Importing the required packages for the needed functionality

mean_mergerx= mean_merger.sort_values("Score", ascending=False)  #Sorting the scores in descending order

mean_mergerx.plot(x="State", y="Score", kind="bar", figsize=(12,6), color = "blue", edgecolor = 'black')  #Bar plot for state and average score
plt.xlabel("State", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.title("Average Score by State", fontsize=18)
plt.grid(False)

plt.show()  #Printing the average score per state
```
